# Подключаем open-cv

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
from mtcnn import MTCNN
```

```python
cv2.__version__
```

```python
path = 'images'
directory = os.listdir(path)
print(directory)
```

```python
detector = MTCNN()
video_capture = cv2.VideoCapture(0)  # webcamera

if not video_capture.isOpened():
    print("Unable to access the camera")
else:
    print("Access to the camera was successfully obtained")

print("Streaming started")
while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    results = detector.detect_faces(rgb)
    for box in results:
        x1, y1, w, h = box['box']
        cv2.rectangle(frame, (x1, y1), (x1+w,y1+h), (0,255,0), 2)
    else:
        print('Нет лиц на фото!')

    # Display the resulting frame
    cv2.imshow("Face detector - to quit press ESC", frame)

    # Exit with ESC
    key = cv2.waitKey(1)
    if key % 256 == 27: # ESC code
        break

# When everything done, release the capture
video_capture.release()
cv2.destroyAllWindows()
print("Streaming ended")
```

```python
ends = ['.jpg', '.png', '.jpeg', '.webp']
photos_rgb = []
for file in directory:
    print(file)
    if os.path.splitext(file)[1] in ends:
        img = cv2.imread(f'images/{file}', cv2.IMREAD_ANYCOLOR)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        detector = MTCNN()
        results = detector.detect_faces(img_rgb)
        # print(results)
        # print('Faces found: ', len(results))
        if len(results) > 0:
            for box in results:
                x1, y1, w, h = box['box']
                cv2.rectangle(img_rgb, (x1, y1), (x1+w,y1+h), (0,255,0), 2)
            else:
                print('Нет лиц на фото!')
            photos_rgb.append(img_rgb)
        else:
            print('Не обнаружены лица на фото!')
    else:
        raise TypeError('Фотография неверного формата!')
```

```python
plt.imshow(photos_rgb[0])
plt.axis('off')
```

```python
image = cv2.imread('images/rem.jpg', cv2.IMREAD_ANYCOLOR)
plt.imshow(image, cmap='gray')
plt.axis('off')
```

```python
# просто смотрю на фотку
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off');
```

```python
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(image_gray, cmap='gray')
plt.axis('off');
```

```python
detector = MTCNN()
results = detector.detect_faces(image_rgb)

print(results)
for box in results:
    x1, y1, w, h = box['box']
    cv2.rectangle(image_rgb, (x1, y1), (x1+w,y1+h), (0,255,0), 2)
else:
    print('Нет лиц на фото!')
plt.imshow(image_rgb)
plt.axis('off');
```

```python
haar_face_cascade = cv2.CascadeClassifier('face_filter/haarcascade_frontalface_default.xml')
faces = haar_face_cascade.detectMultiScale(image_gray, scaleFactor=1.3,
                                           minNeighbors=5,
                                           minSize=(30, 30))
print('Faces found: ', len(faces))
print(faces)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image_rgb, (x, y), (x+w, y+h), (255, 0, 0), 2)
plt.imshow(image_rgb)
plt.axis('off');
```
